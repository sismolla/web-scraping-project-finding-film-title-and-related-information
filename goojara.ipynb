```python
#it takes a full path and returns a dictionary of film title and a full film intro
def goojara_scraper(full_path):
    res = requests.get(full_path)
    soup = BeautifulSoup(res.text, 'html.parser')
    div_page = soup.find('div',class_='marl')

    film_info = {}
    for h1 in div_page.find('h1'):
        title = h1.getText()
        pages = soup.find('div',class_='fimm')
        for index,page in enumerate(pages):
            if index == 0:
                description = page.getText()

                film_info[title] = description 
                print(film_info)
    return film_info

goojara_film = []

#this function will extract every fill address from the the website
def page_no():
    for i in range(1,11):
        page_nos = 'https://www.goojara.to/watch-movies?p=' + str(i)
        base_url = page_nos
        page = requests.get(base_url)
        b_soup = BeautifulSoup(page.text, 'html.parser')
        main = b_soup.select('.dflex a')

        for index , divs in enumerate(main):
            if index >= 7:
                href = divs['href']
                goojara_film.append(goojara_scraper(href))

page_no()

```

```python
#this will save a file as a json file
def save_data(title, data):
    with open(title,'w',encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False,indent=4)

#it loads the data
def load_data(title):
    with open(title,encode='utf-8') as f:
        return json.load(f)
```
